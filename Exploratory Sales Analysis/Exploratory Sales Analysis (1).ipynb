```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate daily sales
daily_sales = train.groupby('date')['sales'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=daily_sales, x='date', y='sales')
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
# Average sales per store
store_sales = train.groupby('store_nbr')['sales'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=store_sales, x='store_nbr', y='sales', palette='viridis')
plt.title("Average Sales per Store")
plt.show()
```

```python
# Average sales per product family
family_sales = train.groupby('family')['sales'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12,5))
sns.barplot(data=family_sales, x='family', y='sales', palette='coolwarm')
plt.title("Average Sales per Product Family")
plt.xticks(rotation=90)
plt.show()
```

```python
train['month'] = train['date'].dt.month
monthly_sales = train.groupby('month')['sales'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=monthly_sales, x='month', y='sales', palette='mako')
plt.title("Monthly Sales Trend")
plt.show()
```
